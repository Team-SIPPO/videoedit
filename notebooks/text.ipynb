# videoedit permit to make text image easy

```python
import videoedit as vedit
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Task: put text on white image.

```python
# generate white image
image = np.ones((1080, 1920, 3), dtype="uint8") * 255
plt.imshow(image)
plt.show()
```

```python
# generate mask
mask = vedit.generate_text_mask(
    img_width=1920,
    img_height=1080,
    text="しっぽにちは",
    text_pos=(1920//2, 1080//2),
    pos="center",
    font_scale=200
)
plt.imshow(mask)
```

```python
# paint chrtacter with mask.
new_image = vedit.fill_color_with_mask(image.copy(), mask, color=(255, 0, 0))
plt.imshow(new_image)
```

```python
# scale_mask
scaled_mask = vedit.scale_and_crop_mask(mask, scale_ratio=2.0)
new_image = vedit.fill_color_with_mask(image.copy(), scaled_mask, color=(0, 255, 0))
plt.imshow(new_image)
```
